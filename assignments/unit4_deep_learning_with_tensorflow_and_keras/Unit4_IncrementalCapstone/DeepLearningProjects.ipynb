## Projects are divided into 4 tasks

### Task 1: Identify customers who churn the bank:

Preprocess the Banking dataset by removing unnecessary features, encoding categorical variables, and creating training and test splits.
Build a Sequential Artificial Neural Network using Keras with suitable activation functions, optimizer, and loss metrics for churn prediction.
Train and evaluate the model using accuracy and confusion matrix metrics to measure prediction performance.

[Data: Churn_Modelling.csv](./data/Churn_Modelling.csv)


### Task 2: Detect humans wearing face masks:

Load and preprocess the image datasets, preparing training, validation, and test data with defined image dimensions.
Develop and train Transfer Learning models using pooling, dropout, and softmax layers, along with callbacks for efficient optimization.
Evaluate and compare model performance on test data, visualizing sample predictions with true and predicted labels to determine the best-performing model.

[Data: Face_Mask_Detection.zip](./data/Face_Mask_Detection.zip)

### Task 3: Classify customer product reviews:

Preprocess the dataset by generating a binary target variable from ratings, tokenizing and padding text data, and dividing it into training and test sets.
Design a CNN-LSTM hybrid model combining convolutional and recurrent layers with embedding, Conv1D, MaxPooling, LSTM, and dense layers for sentiment classification.
Train and evaluate the model using accuracy and loss metrics.

[Data: Grammar and Product Reviews.xlxs](./data/Grammar and Product Reviews.xlsx)

### Task 4: Denoise dirty documents:

Preprocess the dataset by extracting arrays, adding Gaussian noise, and normalizing pixel values.
Build and train an Autoencoder using convolutional and transposed convolutional layers for dental X-ray denoising.
Evaluate performance by visualizing noisy and reconstructed images to assess denoising quality.

[Data: Dental X-Rays.zip](./data/DENTAL_1.NPZ)
